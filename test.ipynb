```python
import requests
from typing import List

DEEPL_API_KEY = "5b2653cc-db5f-4b79-98c9-919648cd63bd:fx"
DEEPL_API_URL = "https://api-free.deepl.com/v2/translate"

HEADERS = {
    "Content-Type": "application/json",
    "Authorization": f"DeepL-Auth-Key {DEEPL_API_KEY}"
}

def translate(text: List[str], target_lang: str="EN") -> str:
    data = {
        "text": text,
        "target_lang": target_lang
    }
    response = requests.post(DEEPL_API_URL, json=data, headers=HEADERS)
    return response.json()['translations'][0]['text']

translate(["안녕"])
```

```python
import requests

url_template = "https://api.upbit.com/v1/candles/minutes/{unit}?market={code}&count={count}"
code = "KRW-BTC"
count = 10
unit = 5

url = url_template.format(unit=unit, code=code, count=count)

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
access_key = os.getenv('UPBIT_OPEN_API_ACCESS_KEY')
secret_key = os.getenv('UPBIT_OPEN_API_SECRET_KEY')
server_url = os.getenv('UPBIT_OPEN_API_SERVER_URL')

upbit = pyupbit.Upbit(access_key, secret_key)
print(upbit.get_balances())
print(upbit.get_balance('KRW'))
```

```python
# limit order bid/ask
# buy_limit_order(ticker, price, volume, contain_req=False)
# sell_limit_order(ticker, price, volume, contain_req=False)
ret = upbit.buy_limit_order("KRW-BTC", 33000000, 1)
print(ret)
```

```python
# market price order
# buy_market_order(ticker, price, contain_req=False) 
# sell_market_order(ticker, price, contain_req=False)
sell = upbit.sell_market_order("KRW-BTC", 1)
print(sell)

# order view/cancel
# view : get_order(ticker_or_uuid, state='wait', kind='normal', contain_req=False)
# cancel : cancel_order(uuid, contain_req=False)
ongoing_order = upbit.get_order("KRW-BTC") # 미체결 주문
finished_order = upbit.get_order("KRW-BTC", state="done") # 완료된 주문
uuid_order = upbit.get_order("UUID") # 특정 주문 상세 조회
```
