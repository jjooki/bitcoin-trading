```python
import requests
from typing import List

DEEPL_API_KEY = "5b2653cc-db5f-4b79-98c9-919648cd63bd:fx"
DEEPL_API_URL = "https://api-free.deepl.com/v2/translate"

HEADERS = {
    "Content-Type": "application/json",
    "Authorization": f"DeepL-Auth-Key {DEEPL_API_KEY}"
}

def translate(text: List[str], target_lang: str="EN") -> str:
    data = {
        "text": text,
        "target_lang": target_lang
    }
    response = requests.post(DEEPL_API_URL, json=data, headers=HEADERS)
    return response.json()['translations'][0]['text']

translate(["안녕"])
```

```python
import requests

url_template = "https://api.upbit.com/v1/candles/minutes/{unit}?market={code}&count={count}"
code = "KRW-BTC"
count = 10
unit = 5

url = url_template.format(unit=unit, code=code, count=count)

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```
